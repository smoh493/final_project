```python
import numpy as np
import pandas as pd
import geopandas as gpd
```

# Read and Clean ZORI File

```python
median_rent = gpd.read_file('medianrent.geojson')
```

```python
median_rent['RegionName'] = median_rent['RegionName'].str.replace(' ', '')
```

```python
median_rent
```

# Read and Clean Population Data

```python
pop_data = pd.read_csv("Combined Data.xlsx - CO-EST2022-HU.csv")
```

```python
pop_data.head()
```

```python
pop_data = pop_data[['Unnamed: 1', 'Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7']]
```

```python
pop_data.rename(columns={'Unnamed: 1': 'RegionName', 'Unnamed: 2': 'Housing Units 2014', 'Unnamed: 3' : 'Population 2014', 'Unnamed: 4' : 'Net Migration 2014', 'Unnamed: 5' : 'Housing Units 2022', 'Unnamed: 6' : 'Population 2022', 'Unnamed: 7' : 'Net Migration 2023'}, inplace=True)
```

```python
pop_data = pop_data.drop_duplicates(subset=['RegionName'])
```

```python
pop_data['RegionName'] = pop_data['RegionName'].str.replace(' ', '')
```

```python
pop_data = pop_data.drop(index=1)
```

```python
pop_data.index = range(1, len(pop_data) + 1)
```

```python
pop_data.head()
```

# Join Files

```python
joined_gdf = median_rent.merge(pop_data, on='RegionName', how='left')
```

```python
joined_gdf.head()
```

# Read and Clean Permit Data

```python
permits = pd.read_csv("permits_clean.csv")
```

```python
permits = permits.drop(columns=['Unnamed: 0'])
```

```python
permits = permits.drop_duplicates(subset=['RegionName'])
```

```python
permits.index = range(1, len(permits) + 1)
```

```python
permits.rename(columns={'STUSAB': 'State'}, inplace=True)
```

```python
permits.head()
```

# Join Files

```python
complete_gdf = joined_gdf.merge(permits, on=['RegionName', 'State'], how='left')
```

```python
complete_gdf.head()
```

# Clean Data

```python
complete_gdf.dropna(inplace=True)
```

```python
complete_gdf.index = range(1, len(complete_gdf) + 1)
```

```python
complete_gdf.rename(columns={'RegionName': 'County', 'Net Migration 2014': 'Net Migration 2015', 'ALL_PERMITS_2014' : 'Permits 2014', 'SINGLE_FAMILY_PERMITS_2014' : 'Single Family Permits 2014', 'ALL_MULTIFAMILY_PERMITS_2014' : 'Multifamily Permits 2014', 'ALL_PERMITS_2022' : 'Permits 2022', 'SINGLE_FAMILY_PERMITS_2022' : 'Single Family Permits 2022', 'ALL_MULTIFAMILY_PERMITS_2022' : 'Multifamily Permits 2022'}, inplace=True)
```

```python
complete_gdf.rename(columns={'2015-01-31': 'Median Rent 2015', '2023-01-31': 'Median Rent 2023'}, inplace=True)
```

```python
complete_gdf['Housing Units 2014'] = complete_gdf['Housing Units 2014'].str.replace(',', '')
complete_gdf['Population 2014'] = complete_gdf['Population 2014'].str.replace(',', '')
complete_gdf['Net Migration 2015'] = complete_gdf['Net Migration 2015'].str.replace(',', '')
complete_gdf['Housing Units 2022'] = complete_gdf['Housing Units 2022'].str.replace(',', '')
complete_gdf['Population 2022'] = complete_gdf['Population 2022'].str.replace(',', '')
complete_gdf['Net Migration 2023'] = complete_gdf['Net Migration 2023'].str.replace(',', '')
```

```python
complete_gdf['Median Rent 2015'] = complete_gdf['Median Rent 2015'].round(2)
complete_gdf['Median Rent 2023'] = complete_gdf['Median Rent 2023'].round(2)
```

```python
complete_gdf['Housing Units 2022'] = complete_gdf['Housing Units 2022'].astype(int)
complete_gdf['Housing Units 2014'] = complete_gdf['Housing Units 2014'].astype(int)
complete_gdf['Net Migration 2023'] = complete_gdf['Net Migration 2023'].astype(int)
complete_gdf['Net Migration 2015'] = complete_gdf['Net Migration 2015'].astype(int)
complete_gdf['Population 2014'] = complete_gdf['Population 2014'].astype(int)
complete_gdf['Population 2022'] = complete_gdf['Population 2022'].astype(int)
complete_gdf['Population 2014'] = complete_gdf['Population 2014'].astype(int)
complete_gdf['Permits 2014'] = complete_gdf['Permits 2014'].astype(int)
complete_gdf['Permits 2022'] = complete_gdf['Permits 2022'].astype(int)
complete_gdf['Single Family Permits 2014'] = complete_gdf['Single Family Permits 2014'].astype(int)
complete_gdf['Multifamily Permits 2014'] = complete_gdf['Multifamily Permits 2014'].astype(int)
complete_gdf['Single Family Permits 2022'] = complete_gdf['Single Family Permits 2022'].astype(int)
complete_gdf['Multifamily Permits 2022'] = complete_gdf['Multifamily Permits 2022'].astype(int)
```

```python
complete_gdf.head()
```

```python
complete_gdf.to_file("clean_data.geojson", driver='GeoJSON')
```
